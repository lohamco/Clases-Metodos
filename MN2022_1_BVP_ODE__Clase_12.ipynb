##Usando odeint()

```python
#programacion funcion dz = f(t,z)
#transformando EDO de orden 2 a
#sistema de EDO de orden 1 (dimension 2)

from scipy import integrate as IT

def Ejemplo1_BFc11(x,Z):
  """ Ejemplo de Burden y Faires
      capitulo 11 ed9
      Ejemplo 1 sec 11.2
      pag. 682
  """
  zo,z1 = Z
  dzo = z1
  dz1 = (1/8)*(32 + 2*x**3 - zo*z1)
  
  dz = np.array([dzo,dz1])
  return dz

## solucion usando odeint


a = 1
b = 3

y_a = 17
y_b = 43/3



X_r = np.linspace(a, b,10) #Discretizacion Dominio
v_prueba = -14 ##valor a determinar para cumplir la condicion de frontera en b
Z_ci = [y_a, v_prueba]

Y_sol = IT.odeint(Ejemplo1_BFc11, Z_ci, X_r, tfirst=True)
                  #por defecto dy/dt = f(y,t) se asume la primer variable
                                          # como la dependiente

##Grafico solucion

plt.plot(X_r, Y_sol[:,0], "*-")
plt.plot(X_r[0],y_a,"ro")
plt.plot(X_r[-1],y_b,"ro")
plt.xlabel("Eje x")
plt.ylabel("y(t)")
plt.grid("on")
plt.show()

Error = abs(y_b - Y_sol[-1,0] )
print("Distancia al y_b: ",Error)
```

##Comparacion Solucion Exacta

```python

Y_exacta = lambda x: x**2 + (16/x)

X_re = np.linspace(1,3,100)
Y_ex = Y_exacta(X_re)

plt.plot(X_re,Y_ex)
plt.plot(X_r, Y_sol[:,0], "*")
plt.legend(["Exacta","Disparo"])
plt.grid("on")
plt.xlabel("Eje x")
plt.show()
```
